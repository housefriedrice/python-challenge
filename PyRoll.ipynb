```python
import os                    # import os and csv packages
import csv
election_list=[]                    # create an empty list to contain each row from csv file as each single list (inside this nested list)

total_votes_for_khan = 0              # set one counter for each of the four candidates
total_votes_for_correy = 0
total_votes_for_li = 0
total_votes_for_otooley = 0


with open('election_data.csv',newline='') as csvfile:                       # open the csv file "election_data.csv"
    
    csvreader = csv.reader(csvfile,delimiter=',')                 # read the csv file
    
    csv_header = next(csvreader)                                # parse each row
    
    for row in csvreader:                                 # loop through the rows
        
        election_list.append(row)                           # add each row as a list to the nested list
        
        if row[2] == "Khan":                                 # add each vote to the counter of the candidate it belongs to
            total_votes_for_khan += 1
        elif row[2] == "Correy":
            total_votes_for_correy += 1
        elif row[2] == "Li":
            total_votes_for_li += 1
        elif row[2] == "O'Tooley":
            total_votes_for_otooley +=1
            
total_votes_for_all = total_votes_for_khan + total_votes_for_correy + total_votes_for_li + total_votes_for_otooley   # sum up to get the total number of votes

fraction_for_khan = total_votes_for_khan / total_votes_for_all

percent_for_khan = "{:.3%}".format(fraction_for_khan)

fraction_for_correy = total_votes_for_correy / total_votes_for_all

percent_for_correy = "{:.3%}".format(fraction_for_correy)

fraction_for_li = total_votes_for_li / total_votes_for_all

percent_for_li = "{:.3%}".format(fraction_for_li)

fraction_for_otooley = total_votes_for_otooley / total_votes_for_all

percent_for_otooley = "{:.3%}".format(fraction_for_otooley)

voteNumberDictionary = {"Khan": total_votes_for_khan,"Correy":total_votes_for_correy,"Li":total_votes_for_li,"O'Tooley":total_votes_for_otooley}

highest_vote_number = max(list(voteNumberDictionary.values()))

winner_name = list(voteNumberDictionary.keys())[list(voteNumberDictionary.values()).index(highest_vote_number)] # use the value to get the key from the dictionary
            
line1 = "Election Results\n"
line2 = "-------------------------\n"
line3 = "Total Votes: " + str(total_votes_for_all) +"\n" 
line4 = "-------------------------\n"
line5 = "Khan: " + str(percent_for_khan) + ' (' + str(total_votes_for_khan) + ')\n'
line6 = "Correy: " + str(percent_for_correy) + ' (' + str(total_votes_for_correy) + ')\n'
line7 = "Li: " + str(percent_for_li) + ' (' + str(total_votes_for_li) + ')\n'
line8 = "O'Tooley: " + str(percent_for_otooley) + ' (' + str(total_votes_for_otooley) + ')\n'
line9 = "-------------------------\n"
line10 = "Winner: " + winner_name +"\n"
line11 = "-------------------------\n"

print(line1+line2+line3+line4+line5+line6+line7+line8+line9+line10+line11)

write_lines_list = [line1,line2,line3,line4,line5,line6,line7,line8,line9,line10,line11]

outputFileDirectory = "output_PyRoll.txt"

f = open(outputFileDirectory,"w+")

f.writelines(write_lines_list)

f.close()

os.startfile(outputFileDirectory)
```
